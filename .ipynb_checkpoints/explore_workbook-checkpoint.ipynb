```python
plots = {1 : [111], 2: [121, 122], 3: [131, 132, 133]}

def countplot(x, y, train):
    
    plt.figure(figsize=(10, 7))
    
    for i, j in enumerate(y):
        plt.subplot(plots[len(y)][i])
        chart = sns.countplot(x=j, hue=x, data=train, alpha=0.75, linewidth=0.5, edgecolor='black', color = 'dodgerblue')
        chart.set_title(j)
        
    return plt.show()
```

```python
crosstab = pd.crosstab(train['job'], train['y_encoded'])

crosstab.plot(kind='bar',stacked=True, figsize=(10,6), color=['#B0B0B0', '#2E8B57'])
plt.xticks(rotation = 45)             
plt.xlabel('Job')
plt.ylabel('Count')
plt.title('Conversion Count By Job')
plt.legend(title = 'Term Deposit', labels = ['No','Yes'])
plt.show()
```

```python
percent_ct = crosstab.div(crosstab.sum(axis=1),axis = 0)

percent_ct.plot(kind='bar',stacked=True, figsize=(10,6), color=['#B0B0B0', '#2E8B57'])
plt.ylim(0, 1.10)
plt.gca().yaxis.set_major_formatter(lambda x, _: f'{x:.0%}')
plt.xticks(rotation = 45, ha= 'right')             
plt.xlabel('Job')
plt.ylabel('Percent')
plt.title('Conversion Rate By Job')
plt.legend(title = 'Term Deposit', labels = ['No','Yes'], bbox_to_anchor=(1.02, 1),   # <-- moves it outside
    loc='upper left')
plt.show()
```

#### Takeaways

- Students, Retired, and Admin seem to have the highest conversions for Term Deposits

```python
def chi_test(train, col):
    
    # set alpha value to 0.05
    alpha = 0.05
    
    # set null and alternative hypothesis 
    null_hypothesis = col + ' and term deposits are independent'
    alternative_hypothesis = col + ' and term deposits are dependent'

    # create an observed crosstab, or contingency table from a dataframe's two columns
    observed = pd.crosstab(train[col], train.y_encoded)

    # run chi-square test
    chi2, p, degf, expected = stats.chi2_contingency(observed)

    # print Null Hypothesis followed by a new line
    print(f'Null Hypothesis: {null_hypothesis}\n')

    # print Alternative Hypothesis followed by a new line
    print(f'Alternative Hypothesis: {alternative_hypothesis}\n')

    # print the chi2 value
    print(f'chi^2 = {chi2}') 

    # print the p-value followed by a new line
    print(f'p     = {p}\n')

    if p < alpha:
        print(f'We reject null hypothesis')
        print(f'There exists some relationship between {col} and bad_resident.')
    else:
        print(f'We fail to reject null hypothesis')
        print(f'There appears to be no significant relationship between {col} and bad_resident.')
        
```

```python
# Let's run a chi squared to compare proportions, to have more confidence
alpha = 0.05
null_hypothesis = "Job Type and Term Deposits are independent"
alternative_hypothesis = "There is a dependency between the job type and term deposit made"

# Setup a crosstab of observed jobs to y
observed = pd.crosstab(train.job, train.y_encoded)

chi2, p, degf, expected = stats.chi2_contingency(observed)

# Assumption check (recommended)
expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
small = (expected_df < 5).sum().sum()
total = expected_df.size

print(f"chi2={chi2:.3f}, degf={degf}, p={p:.6f}")
print(f"Cells with expected < 5: {small}/{total}")


if p < alpha:
    print("Reject the null hypothesis that", null_hypothesis)
    print("Sufficient evidence that", alternative_hypothesis)
else:
    print("Fail to reject the null")
    print("Insufficient evidence to reject the null")
p
```

```python
#Question 2 Does the type of education a client have make it more likely to subscribe?

train.education.value_counts()
```

```python
def plot(train,col1,col2):
    
    crosstab = pd.crosstab(train.col1, train.col2)
    
    per_crosstab = crosstab.div(crosstab.sum(axis=1),axis = 0)
    
    per_crosstab.plot(kind='bar',stacked=True, figsize=(10,6), color=['#B0B0B0', '#2E8B57'])
    plt.ylim(0, 1.10)
    plt.gca().yaxis.set_major_formatter(lambda x, _: f'{x:.0%}')
    plt.xticks(rotation = 45, ha= 'right')             
    plt.xlabel(col1)
    plt.ylabel('Percent')
    plt.title(f'Conversion Rate By {col1}')
    plt.legend(title = 'Term Deposit', labels = ['No','Yes'], bbox_to_anchor=(1.02, 1), loc='upper left') 
    plt.show()
```

```python
ct_edu = pd.crosstab(train.education, train.y_encoded)
```

```python
per_ct_edu = ct_edu.div(ct_edu.sum(axis=1),axis = 0)

per_ct_edu.plot(kind='bar',stacked=True, figsize=(10,6), color=['#B0B0B0', '#2E8B57'])
plt.ylim(0, 1.10)
plt.gca().yaxis.set_major_formatter(lambda x, _: f'{x:.0%}')
plt.xticks(rotation = 45, ha= 'right')             
plt.xlabel('Education Level')
plt.ylabel('Percent')
plt.title('Conversion Rate By Education')
plt.legend(title = 'Term Deposit', labels = ['No','Yes'], bbox_to_anchor=(1.02, 1),   # <-- moves it outside
    loc='upper left')
plt.show()
```

#### Takeaway

- It does not seem to matter the level of education our client has in determining if they will make a term deposit


```python
alpha = 0.05
H_0 = 'Education level and term deposits are independent'
H_a = 'Education level and term deposits are dependent'

observed = pd.crosstab(train.education, train.y_encoded)

chi2, p, degf, expected = stats.chi2_contingency(observed)

# Assumption check (recommended)
expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
small = (expected_df < 5).sum().sum()
total = expected_df.size

print(f"chi2={chi2:.3f}, degf={degf}, p={p:.6f}")
print(f"Cells with expected < 5: {small}/{total}")


if p < alpha:
    print("Reject the null hypothesis that", H_0)
    print("Sufficient evidence that", H_a)
else:
    print("Fail to reject the null")
    print("Insufficient evidence to reject the null")
p
```

```python
def log_test(train, col):
    alpha = 0.05
    H_0 = 'Age has no effect on subscription probability'
    H_a = 'Age affects subscription probability'
    
    X = sm.add_constant(train[col])
    y = train['y_encoded']
    
    model = sm.Logit(y, X).fit()
    
    coef = model.params[col]
    p_value = model.pvalues[col]
    
    print(f"Coefficient: {coef}")
    print(f"P-value: {p_value}")
    
    if p_value < alpha:
        print("Reject the null hypothesis that", H_0)
        print("Sufficient evidence that", H_a)
    else:
        print("Fail to reject the null")
        print("Insufficient evidence to reject the null")
```

```python
alpha = 0.05
H_0 = 'Age has no effect on subscription probability'
H_a = 'Age affects subscription probability'

X = sm.add_constant(train['age'])
y = train['y_encoded']

model = sm.Logit(y, X).fit()
model.summary()
```

```python
coef = model.params['age']
p_value = model.pvalues['age']

print(f"Coefficient: {coef}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis that", H_0)
    print("Sufficient evidence that", H_a)
else:
    print("Fail to reject the null")
    print("Insufficient evidence to reject the null")
p_value
```
